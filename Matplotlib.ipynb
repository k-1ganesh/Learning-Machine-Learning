# Line Graph

```python
age = np.random.randint(10,30,5)
age.sort() 
salary = [30000,35000,37000,35000,40000]

plt.plot(age,salary) # age is on x axis and salary is on y axis.
plt.grid(True)
plt.show() # To disply the plot we need to use it.

# While giving values to x and y axis try to give in ascending order.
```

#### Adding label and title to the plot

```python
plt.plot(age,salary)
plt.xlabel('Ages') # Used to give the x label
plt.ylabel('Salary in Rs') # used to give the y axis label
plt.title('Median Salary by Ages') # used to add the title
plt.grid(True)
plt.show()
```

### Changing and setting edgecolor to black

```python
slices = [60,40]
labeles = ['Sixty','Fourty']
plt.pie(slices,labels = labeles,wedgeprops = {'edgecolor':'black'})
plt.title('Basic Pie Chart')
plt.show()
```

### Changing and adding custom colors to the pie chart

```python
slices = [30,30,20,20]
labeles = ['A','B','C','D']
cust_color  = ['red','green','blue','orange'] # custom colors
plt.pie(slices,labels = labeles,wedgeprops = {'edgecolor':'black'} , colors = cust_color)
plt.title('Basic Pie Chart')
plt.show()
```

### Emphasising on the perticular part of pie chart

```python
# We do this using passing explode list. 
# In explode value 0 means no emphasis and other means emphasis
```

```python
slices = [30,30,20,20]
labeles = ['A','B','C','D']
cust_color  = ['red','green','blue','orange'] # custom colors
explode = [0,0,0,0.1]
plt.pie(slices,labels = labeles,wedgeprops = {'edgecolor':'black'} , colors = cust_color,explode = explode)
plt.title('Basic Pie Chart')
plt.show()
```

### Adding shadow and starting angle to the pie chart

```python
# We can do this using shadow = True
slices = [30,30,20,20]
labeles = ['A','B','C','D']
cust_color  = ['red','green','blue','orange'] # custom colors
explode = [0,0,0,0.1]
plt.pie(slices,labels = labeles,wedgeprops = {'edgecolor':'black'} , colors = cust_color,explode = explode,shadow = True,startangle=90)
plt.title('Basic Pie Chart')
plt.show()
```

### Adding percentage to the each slice of pie chart

```python
# We can do this by passing autopct in pie function
# We can do this using shadow = True
slices = [30,35,20,13]
labeles = ['A','B','C','D']
cust_color  = ['red','green','blue','orange'] # custom colors
# explode = [0,0,0,1]
plt.pie(slices,labels = labeles,wedgeprops = {'edgecolor':'black'} , colors = cust_color,autopct = '%1.2f%%')
plt.title('Basic Pie Chart')
plt.show()
```

# Stack Plot

```python
# Its like line plot but its area is filled
# stackplot(x,y1,y2,y3)
```

```python
minute = [1,2,3,4,5,6]
player1 = [1,3,2,3,4,6]
player2 = [2,1,3,4,2,2]
player3 = [7,2,3,5,6,5]

plt.stackplot(minute,player1,player2,player3)
plt.show()
```

### Adding legend to the stackplot

```python
minute = [1,2,3,4,5,6]
player1 = [1,3,2,3,4,6]
player2 = [2,1,3,4,2,2]
player3 = [7,2,3,5,6,5]

label = ['player1','player2','player3']
plt.stackplot(minute,player1,player2,player3,labels = label)
plt.legend(loc = 'upper left',fontsize='large')
plt.show()

# similarly we can add the colors to the chart as well.
```

# Exploring plt.legend()

```python
# Legend function has interesting parameters
# loc -> used to decide the location of legend
# title -> used to give the title to the legend
# fontsize -> used to control the font size of legend
# framealpha -> used to control the transparency of legend
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot
plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')

# Custom legend
plt.legend(loc='upper right', fontsize='large', title='Trigonometric Functions', shadow=True, fancybox=True, framealpha=0)

plt.title('Sine and Cosine Functions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
age = np.arange(2,20,2)
height = np.random.randint(50,180,9)
weight = np.random.randint(30,80,9)

plt.plot(age,height,label = 'Height') 
plt.plot(age,weight,label = 'Weight')

plt.legend(fontsize = 'large')

plt.fill_between(age,height,alpha = 0.25)

plt.show()
```

```python
# Here plot is filled according to the 0

# We can fill the plot according to the any value.
```

```python
age = np.arange(20,40,2)
company1 = np.random.randint(30000,90000,10)
company2 = np.random.randint(30000,80000,10)

avg_salary = 40000

plt.plot(age,company1,label = 'Company1') 
plt.plot(age,company2,label = 'Company2')

plt.legend(fontsize = 'large')

plt.fill_between(age,company1,avg_salary,alpha = 0.25)
plt.fill_between(age,company2,avg_salary,alpha = 0.25)
plt.show()
```

```python
# Filling according to the conditions
```

```python
age = np.arange(20,40,2)
company1 = np.random.randint(30000,90000,10)
company2 = np.random.randint(30000,80000,10)

avg_salary = 40000

plt.plot(age,company1,label = 'Company1') 
plt.plot(age,company2,label = 'Company2')

plt.legend(fontsize = 'large')

plt.fill_between(age,company1,avg_salary,where=(company1 > avg_salary),alpha = 0.25) # we can apply conditon using where 
plt.fill_between(age,company2,avg_salary,where=(company2 < avg_salary),alpha = 0.25)
plt.show()
```

# Subplot

```python
# Used to create multiple plot on the plot figure.
```

```python
fig , ax = plt.subplots()
# It returns 2 things figure and axis. 

age1 = [ 10,12,14,16]
salary1 = [500,450,600,1000]
ax.bar(age1,salary1,color='r',width = 0.5)
ax.set_title('Salary / Day')
ax.set_xlabel('Age')
ax.set_ylabel('Salary')

plt.show()
```

```python
# subplots(nrows , ncols )  --> Used to set how many plots we want in 1 figure.

# Lets try to plot 2 graph in 1 figure.

fig , (ax1,ax2) = plt.subplots(nrows = 2 , ncols = 1)

age1 = [ 10,12,14,16]
salary1 = [500,450,600,1000]

ax1.bar(age1,salary1,color='r',width = 0.5,label = 'Bar Graph')
ax1.set_title('Salary / Day')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.legend()

ax2.plot(age1,salary1,marker='o',label='line Graph')
ax2.set_ylabel('Salary')
ax2.set_xlabel('Age')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# subplots(nrows , ncols )  --> Used to set how many plots we want in 1 figure.

# Lets try to plot 2 graph in 1 figure.

fig , (ax1,ax2) = plt.subplots(nrows = 2 , ncols = 1,sharex = True)

age1 = [ 10,12,14,16]
salary1 = [500,450,600,1000]

ax1.bar(age1,salary1,color='r',width = 0.5,label = 'Bar Graph')
ax1.set_title('Salary / Day')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.legend()

ax2.plot(age1,salary1,marker='o',label='line Graph')
ax2.set_ylabel('Salary')
ax2.set_xlabel('Age')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Getting both plot on different figures.

age1 = [ 10,12,14,16]
salary1 = [500,450,600,1000]

fig1 , ax1 = plt.subplots()
fig2 , ax2 = plt.subplots()

ax1.bar(age1,salary1,color='r',width = 0.5,label = 'Bar Graph')
ax1.set_title('Salary / Day')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.legend()

ax2.plot(age1,salary1,marker='o',label='line Graph')
ax2.set_ylabel('Salary')
ax2.set_xlabel('Age')
ax2.legend()

plt.tight_layout()
plt.show()
```
