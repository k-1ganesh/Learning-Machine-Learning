```python
df = sns.load_dataset("penguins")
df
```

```python
sns.lineplot(x = 'bill_length_mm',y = 'body_mass_g' , data = df.head(100).dropna(),marker='o')
plt.show()
```

```python
# Now here in above plot contain both male female responce.
# We can certainly figure out whichh is male and which is female.

We Can Exphasis plot based on One of the column of dataFrame.
We can do this using hue parameter
```

```python
sns.lineplot(x = 'bill_length_mm',y = 'body_mass_g' , data = df.dropna(),marker='o' , hue = 'sex')
plt.show()

# Now we can easily do analysis as graph is seperated according to the sex.
```

```python
# I can achive the same thing using pandas and matplotlib. 
# For that i need to seperate the data according to the gender.
# Then i need to plot 2 lineplots with blue and orange color.
df1 = df.dropna()
df1  = df1.sort_values(by = 'bill_length_mm')
group = df1.groupby('sex')
x1 = group.get_group('Male')['bill_length_mm']
y1 = group.get_group('Male')['body_mass_g']

x2 = group.get_group('Female')['bill_length_mm']
y2 = group.get_group('Female')['body_mass_g']


plt.plot(x1,y1,marker='o')
plt.plot(x2,y2,marker='o')
plt.show()


# Its a same graph but very lengthy code. 
# seaborn is lovely ;)
```

```python
# When plot has different group formed using hue then we can use style to diffeentiat drawing style of different groups.
sns.lineplot(x = 'bill_length_mm',y = 'body_mass_g' , data = df.dropna(),marker='o' , hue = 'sex' , style = 'sex')
plt.show()
```

## Changing color and marker style for different groups

```python
# When plot has different group formed using hue then we can use style to diffeentiat drawing style of different groups.
sns.lineplot(x = 'bill_length_mm',y = 'body_mass_g' , data = df.dropna(),marker='o' , hue = 'sex' , style = 'sex',palette='Accent')
plt.show()
```

```python
# Adding custom colors using palette
sns.lineplot(x = 'bill_length_mm',y = 'body_mass_g' , data = df.dropna(),marker='o' , hue = 'sex' , style = 'sex',palette = ['#53cedb','#72e885'])
plt.show()
```

# Bar Plot


```python
# Bar plot is used to plot the catogorical data on the x axis .
# By default bar plot plots the mean on the y axis.
# Statistical function can be choosed using estimator attribute
```

```python
df = sns.load_dataset('penguins')
```

```python
df
```

```python
sns.barplot(x = 'island',y = 'bill_length_mm',data = df)
plt.show()
```

```python
sns.barplot(x = 'island',y = 'bill_length_mm',data = df, estimator = np.max , ci = 'sd')
# Errorbar indicates standard deviation.
# Now Y axis has max bill length for each catogory.
plt.show()
```

```python
sns.barplot(x = 'island',y = 'bill_length_mm',data = df,hue = 'sex',width = 0.5)
plt.show()
```

```python
# I can change the order of the bars as well.

order  = [ 'Dream' , 'Torgersen' , 'Biscoe' ] 
sns.barplot(x = 'island',y = 'bill_length_mm',data = df , order = order , hue = 'sex')
plt.show()
```

```python
# I can even change the hue order.
sns.barplot(x = 'island',y = 'bill_length_mm',data = df , order = order , hue = 'sex', hue_order = ['Female','Male'],errorbar = None,width = 0.6,)
plt.legend(loc = "lower left")
plt.show()
```

```python
# I can plot horizontal barplot using orient = 'h'

sns.barplot(x = 'flipper_length_mm' , y = 'bill_length_mm' , data = df.head(10) , orient = 'h',palette = ['#72e8e0'],width = 0.5)
plt.show()
```

# Scatter Plot 

```python
df = sns.load_dataset('penguins')
df
```

```python
# Lets check relation between bill length and body mass using scatter plot.

sns.scatterplot(x = 'bill_length_mm' , y = 'body_mass_g' , data  = df)
plt.show()
```

```python
# lets differentiate data by sex.
sns.scatterplot(x = 'bill_length_mm' , y = 'body_mass_g' , data  = df , hue = 'sex')
plt.show()
```

```python
# We can adjust size according to the size of data . 
# Lets we want size of each dot should depend on body mass . Higher the size and higher the mass.
sns.scatterplot(x = 'bill_length_mm' , y = 'body_mass_g' , data  = df , hue = 'sex' , size = 'body_mass_g')
# Bigger dots indicates the higher body mass
plt.show()
```
