### dropna()

```python
# Dropna() is used to delete the rows which has null values
```

```python
# Defalut Dropna() will delete all the rows which has atlest 1 null value

# dropna( axis , how) 
# axis -> to delete row or column
# How -> to specify whether 1 value or all value. ( any , all )
```

```python
df1
```

```python
df1.dropna() # deleted all the rows which has Nan
```

```python
df1.dropna(how='all') # Rows will be deleted iff all values are na 
```

```python
df1.dropna(axis = 'columns') # Deleting columns which has atlest 1 null value.
```

```python
# Deleting rows by checking whether null is in specific column
# for that we will use (subset)
df1.dropna(how='any',subset=['D']) # Deleted only that row in which D is null
```

```python
# Deleting only those rows in which both C and D are null
df1.dropna(how='all' , subset=['C','D']) 
# In this all will check na presence in both C and D
```
